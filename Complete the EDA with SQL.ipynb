## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
query = 'SELECT DISTINCT "Launch_Site" FROM SPACEXTBL;'
unique_launch_sites = pd.read_sql_query(query, con)
unique_launch_sites
```

```python
query = 'SELECT DISTINCT "Landing_Outcome" FROM SPACEXTBL;'
unique_launch_sites = pd.read_sql_query(query, con)
unique_launch_sites
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
query = 'SELECT * FROM SPACEXTBL WHERE "Launch_Site" LIKE "CCA%" LIMIT 5;'
launch_sites_with_prefix_CCA = pd.read_sql_query(query, con)
launch_sites_with_prefix_CCA
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
query = 'SELECT SUM("PAYLOAD_MASS__KG_") AS TotalPayloadMass FROM SPACEXTBL WHERE "Customer" LIKE "%NASA (CRS%";'
total_payload_mass_nasa_crs = pd.read_sql_query(query, con)
total_payload_mass_nasa_crs
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
query = 'SELECT AVG("PAYLOAD_MASS__KG_") AS AvgPayloadMass FROM SPACEXTBL WHERE "Booster_Version" LIKE "%F9 v1.1%";'
AvePayloadMass= pd.read_sql_query(query, con)
AvePayloadMass
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
query = 'SELECT MIN("Date") AS Earliest FROM SPACEXTBL WHERE "Mission_Outcome" LIKE "Success";'
first_succesful_landing = pd.read_sql_query(query, con)
first_succesful_landing
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
query = 'SELECT "Booster_Version" FROM SPACEXTBL WHERE "Landing_Outcome" = "Success (drone ship)" AND "PAYLOAD_MASS__KG_" > 4000 AND "PAYLOAD_MASS__KG_" < 6000;'
boosters_success_drone_ship = pd.read_sql_query(query, con)
boosters_success_drone_ship
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
query = 'SELECT "Mission_Outcome", COUNT(*) AS Count  WHERE "Mission_Outcome" LIKE "%Success%";'
mission_outcomes = pd.read_sql_query(query, con)
mission_outcomes
```

```python
query = 'SELECT "Mission_Outcome", COUNT(*) AS Count FROM SPACEXTBL GROUP BY "Mission_Outcome";'
mission_outcomes = pd.read_sql_query(query, con)
mission_outcomes
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
query = 'SELECT "Booster_Version" FROM SPACEXTBL WHERE "Payload_Mass__kg_" = (SELECT MAX("Payload_Mass__kg_") FROM SPACEXTBL);'
booster_versions = pd.read_sql_query(query, con)
booster_versions
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
query = 'SELECT strftime("%m", Date) AS Month, "Landing_Outcome", "Booster_Version", "Launch_Site" FROM SPACEXTBL WHERE substr(Date, 7, 4) = "%2015%" AND "Landing_Outcome" LIKE "%Failure%" AND "Booster_Version" LIKE "%F9%" AND "LaunchSite" LIKE "%Drone Ship%";'
records_2015 = pd.read_sql_query(query, con)
records_2015
```

### Task 10




##### Rank the  count of  successful landing_outcomes between the date 04-06-2010 and 20-03-2017 in descending order.


```python


query = 'SELECT "Landing_Outcome", COUNT(*) AS Count FROM SPACEXTBL WHERE date(substr(Date, 4, 2) || "-" || substr(Date, 1, 2) || "-" || substr(Date, 7, 4)) BETWEEN date("2010-06-04") AND date("2017-03-20") AND "Landing_Outcome" LIKE "%Success%" GROUP BY "Landing_Outcome" ORDER BY Count DESC'
ranking = pd.read_sql_query(query, con)
ranking
```
